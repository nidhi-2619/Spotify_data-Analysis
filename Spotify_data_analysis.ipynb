```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_tracks = pd.read_csv('/content/tracks.csv')
spotify_tracks.head()
```

```python
# null values in the data sets
#to check if there are any null data
pd.isnull(spotify_tracks).sum() 
#sum will calculate the number of null values
```

```python
# info method will tell about the number of rows and columns in the dataset
spotify_tracks.info()
# it will also check the datatype and memory usage
```

```python
#we are tring to find the least favourite or least popular song 
# head is used to get only 10 values which are least favourite
sorted_tracks = spotify_tracks.sort_values('popularity',ascending =True).head(10)
sorted_tracks
```

```python
#statistic analysis of the data
spotify_tracks.describe().transpose()
```

```python
#we are finding most popular songs
# we have set inplace to false so that our original data does not change
most_popular = spotify_tracks.query('popularity>90',inplace = False).sort_values('popularity',ascending=False)
#we are using slice operator to show only top 10 song , so we do not use slice it will so all the song 
#from most popular to least popular
most_popular[:10]
```

```python
# we are changing the index with release date and we have set the inplace true so that the changes reflect 
# original data
spotify_tracks.set_index("release_date",inplace = True)
spotify_tracks.index=pd.to_datetime(spotify_tracks.index)
spotify_tracks.head()
```

```python
# we will convert the duration from milliseconds to seconds
spotify_tracks['duration']=spotify_tracks['duration_ms'].apply(lambda x: round(x/1000))
# we have droped the duration and axis is used to specify that duration ms is column
spotify_tracks.drop('duration_ms',inplace=True,axis = 1)
```

```python
# here we are using pearson method for finding correlation of spotify data
spotify_corr = spotify_tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(spotify_corr,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='viridis',linewidth=1,linecolor='Black')
heatmap.set_title('Correlation Heatmap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
# for regression plot
# we are taking only 4% of our data and try to plot the 2 regression using this data
filtered_data = spotify_tracks.sample(int(0.004*len(spotify_tracks)))
```

```python
print(len(filtered_data))
```

```python
# creating regression plot between loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=filtered_data,y='loudness',x='energy',scatter_kws={'color': 'palevioletred'}, line_kws={"color": "red"}).set(title='Loudness Vs Energy Correlation')
```

```python
# creating the regression plot between popularity and acoustic
plt.figure(figsize=(10,6))
sns.regplot(data=filtered_data,y='popularity',x='acousticness', line_kws={"color": "red"}).set(title='Popularity Vs Acoustic Correlation')
```

```python
# we are creating a new column year from the release date
spotify_tracks['dates']=spotify_tracks.index.get_level_values('release_date')
spotify_tracks.dates=pd.to_datetime(spotify_tracks.dates)
years=spotify_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='Number of songs per year')
```

```python
# duration over the years
total_duration = spotify_tracks.duration
fig_dims = (18,7)
fig,ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years,y=total_duration,ax=ax,errwidth=False).set(title='Year Vs Duration')
plt.xticks(rotation=90)
```
